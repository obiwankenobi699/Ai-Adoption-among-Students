# Data Cleaning

This notebook prepares the dataset for analysis.

Steps:
- Remove duplicates
- Handle missing values
- Convert date format
- Create new features
- Save cleaned dataset

```python
import os
import pandas as pd

input_path = 'assets/ai_assistant_usage_student_life.csv'
output_path = 'assets/cleaned_ai_usage.csv'

if os.path.exists(output_path):
    print('Cleaned dataset already exists. Skipping cleaning.')
else:
    print('Cleaning dataset...')

    df = pd.read_csv(input_path)

    # Remove duplicates
    df = df.drop_duplicates()

    # Handle missing values
    for col in df.select_dtypes(include=['number']):
        df[col] = df[col].fillna(df[col].median())

    # Convert date
    if 'SessionDate' in df.columns:
        df['SessionDate'] = pd.to_datetime(df['SessionDate'], errors='coerce')

    # Feature engineering
    df['HighUsage'] = df['TotalPrompts'] > df['TotalPrompts'].median()
    df['Efficiency'] = df['TotalPrompts'] / df['SessionLengthMin']

    if 'SessionDate' in df.columns:
        df['Month'] = df['SessionDate'].dt.month

    # Save cleaned dataset
    df.to_csv(output_path, index=False)

    print('Cleaned dataset saved to:', output_path)

df.head()
```
